```python
import plotly.graph_objs as go

def make_graph(stock_data, stock, title):
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=stock_data['Date'], y=stock_data['Close'], name=stock))
    fig.update_layout(title=title, xaxis_title='Date', yaxis_title='Close Price (USD)')
    fig.show()

make_graph(tesla_data, 'Tesla', 'Tesla Stock Price History')
```

```python
make_graph(tesla_data, 'Tesla', 'Tesla Stock Price History')
```

```python
make_graph(gme_data, 'GameStop', 'GameStop Stock Price History')
```
